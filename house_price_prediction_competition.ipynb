### As we can see,the variables are somewhat normally distributed,hence we will apply log transformation later on,so that the algorithm can work on the data in a better manner.

```python
for feature in continuous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
for feature in numerical_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
#log of 0 is not defined
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df,df['SalePrice'],test_size=0.1,random_state=0)

X_train.shape, X_test.shape
```

```python
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,df[feature].isnull().sum()*100/len(df[feature])))
```

```python
## Replace missing value with a new value
def replace_cat_feature(df,features_nan):
    data=df.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df=replace_cat_feature(df,features_nan)
```

```python
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
numerical_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,df[feature].isnull().sum()*100/len(df[feature])))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

```python
df.head(50)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df[feature]=df['YrSold']-df[feature]

df.head()
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df[num_features]
```

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature).count()/len(df[feature])
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df.head(100)
```

```python
data=pd.get_dummies(df)
data
```

```python
X=data.drop(['Id','SalePrice'],axis=1)
y=data['SalePrice']
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
lasso_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
lasso_model.fit(X,y)
```

```python
lasso_model.get_support()
```

```python
important_features = X.columns[(lasso_model.get_support())]
print('total features: {}'.format((X.shape[1])))
print('selected features: {}'.format(len(important_features)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(lasso_model.estimator_.coef_ == 0)))
```

```python
X=X[important_features]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(X_train,y_train)
```

```python
yhat=linear.predict(X_test)
print("The accuracy is:",mean_absolute_error(y_test,yhat))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(X_train,y_train)
```

```python
rf_yhat=random_forest.predict(X_test)
print("THe accuracy is:",mean_absolute_error(y_test,rf_yhat))
```
